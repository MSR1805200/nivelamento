```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/gabarito_nivelamento/aula_02_exemplo_01.csv")

df = df.rename(columns={
    "NY.GDP.PCAP.CD": "GDP_PC",
    "SL.TLF.ADVN.ZS" : "forca_trab_educ",
    "AG.LND.ARBL.HA.PC": "arable_land",
    "NV.IND.TOTL.ZS": "industria_PERCPIB",
    "SE.XPD.TOTL.GD.ZS": "gasto_educ_PERCPIB",
    "NE.GDI.FTOT.CD": "FBKF",
    "SP.POP.TOTL": "populacao"})

df
```

# E se quiséssemos saber qual a população total em cada um dos quintis de riqueza?


```python
df["quintil"] = pd.qcut(df["GDP_PC"], q = 5, labels = ["pobres", "médio pobres", "médio", "médio ricos", "ricos"])
df
```

```python
df.groupby("quintil")["populacao"].sum()
```

# E se quiséssemos saber qual é o investimento médio em educação comparando por quintis?


```python
df.groupby("quintil")["gasto_educ_PERCPIB"].mean()
```

# Qual o investimento mínimo, máximo, a mediana e a média comparando por quintis?


```python
df.groupby("quintil")["gasto_educ_PERCPIB"].agg(["count","min","max", "median","mean"])
```
