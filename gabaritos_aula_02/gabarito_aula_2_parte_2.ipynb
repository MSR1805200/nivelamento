### Crie um DataFrame chamado df_amostra contendo apenas 5 mulheres selecionadas aleatoriamente.


```python
df_amostra = df.sample(5, replace = False)
```

### Se tentarmos unir essa amostra de 5 pessoas com a base de PIB (que tem 27 estados), qual comportamento você espera observar ao testar:


```python
df_pib_pc = pd.DataFrame({'PIB_PC': {'SP': 24456, 'MG': 14232, 'RJ': 21621, 'RS': 18378, 'PR': 16927,
                                     'BA':  8378, 'CE':  7111, 'SC': 20368, 'PE':  8064, 'GO': 12878,
                                     'DF': 45977, 'PA':  7992, 'MA':  6103, 'PB':  6865, 'PI':  5372,
                                     'ES': 20230, 'RN':  8202, 'AL':  6227, 'AM': 14014, 'SE':  9778,
                                     'MS': 14188, 'MT': 17927, 'AC':  9896, 'TO': 10223, 'RO': 11976,
                                     'AP': 11032, 'RR': 11844}}).reset_index(names='UF')
df_pib_pc
```

### Left Join: Veremos as 5 mulheres com seus PIBs ou os 27 estados com PIBs vazios?


```python
df_left = pd.merge(df_amostra, df_pib_pc, how = 'left', left_on= "uf", right_on = "UF")
df_left
```

### Right Join: O que acontece com os estados que não possuem nenhuma representante nessas 5 mulheres selecionadas?


```python
df_right = pd.merge(df_amostra, df_pib_pc, how="right", left_on = "uf", right_on= "UF" )
df_right
```
